```python
import pandas as pd
```

```python
data = [[3, 108939], [2, 12747], [8, 87709], [6, 91796]]
accounts = pd.DataFrame(data, columns=['account_id', 'income']).astype({'account_id':'Int64', 'income':'Int64'})
```

```python
accounts
```

```python
low = 0
avg = 0
high = 0
for i in accounts['income']: 
    if i < 20000: 
        low += 1
    elif i >= 20000 and i <= 50000: 
        avg += 1
    else: 
        high += 1
```

```python
df = pd.DataFrame([['Low Salary', low], ['Average Salary', avg], ['High Salary', high]], columns=['category', 'accounts_count'])
```
